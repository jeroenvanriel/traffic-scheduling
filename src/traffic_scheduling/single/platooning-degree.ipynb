### Distribution visualization

Plot some route arrival sequences to give an idea of high and low degrees of platooning in the arrival distribution.

```python
from single.basics import bimodal_exponential, generate_instance

F_high = bimodal_exponential(p=0.5, s1=0.1, s2=10.00)
F_low = bimodal_exponential(p=0.1, s1=0.1, s2=5.6)

options = {
    'block_text': False,
    'no_axis': True,
}
# number of arriving vehicles
n = 20
```

```python
generate_instance(F_high, n=n, R=1).visualize(**options)
generate_instance(F_high, n=n, R=1).visualize(**options)
generate_instance(F_high, n=n, R=1).visualize(**options)
generate_instance(F_high, n=n, R=1).visualize(**options)
generate_instance(F_high, n=n, R=1).visualize(**options)
generate_instance(F_high, n=n, R=1).visualize(**options)
generate_instance(F_high, n=n, R=1).visualize(**options)
generate_instance(F_high, n=n, R=1).visualize(**options)
generate_instance(F_high, n=n, R=1).visualize(**options)
```

```python
generate_instance(F_low, n=n, R=1).visualize(**options)
generate_instance(F_low, n=n, R=1).visualize(**options)
generate_instance(F_low, n=n, R=1).visualize(**options)
generate_instance(F_low, n=n, R=1).visualize(**options)
generate_instance(F_low, n=n, R=1).visualize(**options)
generate_instance(F_low, n=n, R=1).visualize(**options)
generate_instance(F_low, n=n, R=1).visualize(**options)
generate_instance(F_low, n=n, R=1).visualize(**options)
generate_instance(F_low, n=n, R=1).visualize(**options)
```
